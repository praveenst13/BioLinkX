```python
import pandas as pd
import torch
import torch.nn.functional as F
from torch_geometric.data import HeteroData
from torch_geometric.nn import HeteroConv, SAGEConv, GATConv
from torch_geometric.transforms import RandomLinkSplit
from sklearn.metrics import roc_auc_score, average_precision_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import warnings
import os
import itertools
from collections import defaultdict
import csv
import sys # For clean exit

# Suppress warnings for cleaner output
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

# --- CONFIGURATION ---
FEATURE_DIM = 32 # The target dimensionality for all node features
LEARNING_RATE = 0.005
NUM_EPOCHS = 1
 # Increased prediction output
# Set up device: Ensure Tensors created below are moved to this device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# --- SAVE CONFIGURATION ---
SAVE_DIR = '.'
os.makedirs(SAVE_DIR, exist_ok=True)
MODEL_ENCODER_PATH = os.path.join(SAVE_DIR, 'best_hgnn_encoder_omics.pt')
MODEL_PREDICTOR_PATH = os.path.join(SAVE_DIR, 'best_link_predictor_omics.pt')
# --- END CONFIGURATION ---

# --- FILE MAPPING (CRITICAL: UPDATE THESE PATHS!) ---
# NOTE: Assuming these paths are correct as provided by the user
FILE_MAP = {
    "disease to disease.tsv": "/content/drive/MyDrive/biosnap/DD-Miner_miner-disease-disease.tsv",
    "disease to drug tsv": "/content/drive/MyDrive/biosnap/DCh-Miner_miner-disease-chemical.tsv",
    "disease to function.tsv": "/content/drive/MyDrive/biosnap/DF-Miner_miner-disease-function.tsv",
    "disease to gene-1 tsv": "/content/drive/MyDrive/biosnap/DG-AssocMiner_miner-disease-gene.tsv",
    "disease to gene-2 tsv": "/content/drive/MyDrive/biosnap/DG-Miner_miner-disease-gene.tsv",
    "drug to drug tsv": "/content/drive/MyDrive/biosnap/ChCh-Miner_durgbank-chem-chem.tsv",
    "drug to drug to sideeffects.csv": "/content/drive/MyDrive/biosnap/ChChSe-Decagon_polypharmacy.csv",
    "drug to gene interaction .tsv": "/content/drive/MyDrive/biosnap/ChG-Miner_miner-chem-gene.tsv",
    "drug to sideeffects csv": "/content/drive/MyDrive/biosnap/ChSe-Decagon_monopharmacy.csv",
    "function to function.tsv": "/content/drive/MyDrive/biosnap/FF-Miner_miner-func-func.tsv",
    "gene to function tsv": "/content/drive/MyDrive/biosnap/GF-Miner_miner-gene-function.tsv",
    "gene to protein tsv": "/content/drive/MyDrive/biosnap/GP-Miner_miner-gene-protein.tsv",


  # OMICS Files (Now uses uploaded files if accessible)
    "transcriptomics.soft": "/content/drive/MyDrive/biosnap/transcriptomics.soft",
    "epigenomics.soft": "/content/drive/MyDrive/biosnap/epigenomics.soft",
    "proteomics.txt": "/content/drive/MyDrive/biosnap/proteomics.txt",
    "liver.txt": "/content/drive/MyDrive/biosnap/liver.txt",
    "kidney.txt": "/content/drive/MyDrive/biosnap/kidney.txt",
    "serum.txt": "/content/drive/MyDrive/biosnap/serum.txt",
}
# Define all relationships based on your requested structure
RELATION_DEFINITIONS = [
    ("disease to disease.tsv", 'disease', 'disease', 'similar_to', [0, 1, '\t']),
    ("drug to drug tsv", 'drug', 'drug', 'interacts_with', [0, 1, '\t']),
    ("function to function.tsv", 'function', 'function', 'related_to', [0, 1, '\t']),

    # Disease-Gene: DG-1 uses columns 0 (Disease ID) and 2 (Gene ID)
    ("disease to gene-1 tsv", 'disease', 'gene', 'associated_with_1', [0, 2, '\t']),
    # Disease-Gene: DG-2 uses columns 0 (Disease) and 1 (Gene)
    ("disease to gene-2 tsv", 'disease', 'gene', 'associated_with_2', [0, 1, '\t']),

    # Other Heterogeneous Links
    ("disease to drug tsv", 'disease', 'drug', 'treated_by', [0, 1, '\t']),
    ("disease to function.tsv", 'disease', 'function', 'affects', [0, 1, '\t']),
    ("drug to gene interaction .tsv", 'drug', 'gene', 'targets', [0, 1, '\t']),
    ("gene to protein tsv", 'gene', 'protein', 'codes_for', [0, 1, '\t']),
    ("gene to function tsv", 'gene', 'function', 'performs', [2, 0, '\t']), # Note: reversed indices based on common file format
    ("drug to sideeffects csv", 'drug', 'side_effect', 'causes', [0, 1, ',']),
]
# Polypharmacy (Drug-Drug and Drug-SideEffect)
POLYPHARMACY_DRUG_DRUG = ("drug to drug to sideeffects.csv", 'drug', 'drug', 'interacts_poly', [0, 1, 2, ','])
POLYPHARMACY_DRUG_SE = ("drug to drug to sideeffects.csv", 'drug', 'side_effect', 'causes_poly', [0, 1, 2, ','])

# --- 1. NODE ID MAPPING SETUP ---
def load_and_collect_ids(filename_key, type_A, type_B, col_A, col_B, sep):
    """Loads file and collects unique node IDs for two node types with robust column check and TSV parsing."""
    path = FILE_MAP.get(filename_key)
    is_uploaded = isinstance(path, str) and path.startswith("uploaded:")
    if not is_uploaded and (path is None or not os.path.exists(path)):
        if filename_key != "disease to gene-2 tsv":
            print(f"Warning: File not found at {path}. Skipping relation for {filename_key}.")
        return set(), set()
    try:
        # FIX 1: Use quoting=csv.QUOTE_NONE to correctly parse TSV files with quoted fields
        quoting_mode = csv.QUOTE_NONE if sep == '\t' else csv.QUOTE_MINIMAL
        df = pd.read_csv(path, sep=sep, skiprows=1, header=None, comment='#', engine='python',
                         skipinitialspace=True, quoting=quoting_mode)
        # FIX: Robust check for column bounds
        max_col_index = max(col_A, col_B)
        if max_col_index >= df.shape[1]:
            print(f"Error: Columns {col_A} or {col_B} out of bounds for {filename_key} (DataFrame has {df.shape[1]} columns). Skipping.")
            return set(), set()
        idsA = set(df.iloc[:, col_A].astype(str).str.strip().unique())
        idsB = set(df.iloc[:, col_B].astype(str).str.strip().unique())
        return {i for i in idsA if i and i != 'nan'}, {i for i in idsB if i and i != 'nan'}
    except Exception as e:
        print(f"Error processing {filename_key}: {e}")
        return set(), set()

# Initialize node sets and collect from network definitions
all_nodes = {
    'disease': set(), 'drug': set(), 'function': set(), 'gene': set(), 'protein': set(), 'side_effect': set(), 'metabolite': set()
}
print("--- Collecting Node IDs from Network Files ---")
for filename, typeA, typeB, _, (colA, colB, sep) in RELATION_DEFINITIONS:
    idsA, idsB = load_and_collect_ids(filename, typeA, typeB, colA, colB, sep)
    all_nodes[typeA].update(idsA)
    all_nodes[typeB].update(idsB)

filename_poly, typeA, typeB, _, (colA, colB, colC, sep) = POLYPHARMACY_DRUG_SE
path_poly = FILE_MAP.get(filename_poly)
is_poly_uploaded = isinstance(path_poly, str) and path_poly.startswith("uploaded:")
try:
    if path_poly and (is_poly_uploaded or os.path.exists(path_poly)):
        df_poly = pd.read_csv(path_poly, sep=sep, skiprows=1, header=None, comment='#', engine='python', skipinitialspace=True)
        max_col_index = max(colA, colB, colC)
        if max_col_index >= df_poly.shape[1]:
            print(f"Error: Columns out of bounds for {filename_poly} (DataFrame has {df_poly.shape[1]} columns).")
        else:
            all_nodes['drug'].update(df_poly.iloc[:, colA].astype(str).str.strip().unique())
            all_nodes['drug'].update(df_poly.iloc[:, colB].astype(str).str.strip().unique())
            all_nodes['side_effect'].update(df_poly.iloc[:, colC].astype(str).str.strip().unique())
except Exception as e:
    print(f"Error processing {filename_poly}: {e}")

node_maps = {}
node_id_to_string = defaultdict(list)
for node_type, ids in all_nodes.items():
    clean_ids = {i for i in ids if i and i != 'nan'}
    categories = pd.Categorical(list(clean_ids)).categories
    node_maps[node_type] = pd.Series(np.arange(len(categories)), index=categories)
    node_id_to_string[node_type] = categories.tolist()
    if node_type != 'metabolite' or len(categories) > 0:
        print(f"Found {len(categories)} unique {node_type} nodes.")

# --- 2. OMICS FEATURE GENERATION FUNCTIONS (Gene/Protein Only) ---
def process_features_with_pca(df, feature_cols, target_dim=FEATURE_DIM):
    """Standardizes, fills NaNs, and applies PCA."""
    if df.empty or len(feature_cols) == 0:
        return pd.DataFrame()
    X = df[feature_cols].copy()
    X = X.fillna(X.mean()).replace([np.inf, -np.inf], np.nan).fillna(0)
    n_samples = X.shape[0]
    n_features = X.shape[1]

    if n_samples <= 1:
        return pd.DataFrame(np.zeros((n_samples, target_dim)), index=X.index)

    n_components = min(target_dim, n_features, n_samples - 1)
    if n_components <= 0:
        return pd.DataFrame(np.zeros((n_samples, target_dim)), index=X.index)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X_scaled)

    if X_pca.shape[1] < target_dim:
        padding = np.zeros((X_pca.shape[0], target_dim - X_pca.shape[1]))
        X_pca = np.hstack([X_pca, padding])

    return pd.DataFrame(X_pca, index=X.index, columns=[f'feat_{i}' for i in range(target_dim)])

def load_transcriptomics_features():
    path = FILE_MAP.get("transcriptomics.soft")
    if not isinstance(path, str) or not (path.startswith("uploaded:") or os.path.exists(path)): return pd.DataFrame()
    try:
        # Transcriptomics data often starts with '!' comments, so we use comment='!'
        df = pd.read_csv(path, sep='\t', comment='!', engine='python', encoding='latin1')
        print(f"Transcriptomics: Raw data shape: {df.shape}")
        expr_cols = [col for col in df.columns if col.startswith('GSM')]
        # Identify the Gene ID column
        gene_id_col = next((c for c in df.columns if 'Gene' in c and 'ID' in c), None)

        if not expr_cols:
            print("Transcriptomics: Could not find any 'GSM' expression columns.")
            return pd.DataFrame()
        if not gene_id_col:
            print("Transcriptomics: Could not find a 'Gene ID' column.")
            return pd.DataFrame()

        df = df.set_index(gene_id_col)[expr_cols].apply(pd.to_numeric, errors='coerce')
        # Clean up index and drop rows where all values are NaN
        df = df.loc[df.index.dropna()]
        df.index = df.index.astype(str).str.strip()
        df = df.loc[df.index != '0']
        df.dropna(how='all', inplace=True)
        # NEW FIX: Ensure unique index by taking the mean of duplicate Gene IDs
        if not df.index.is_unique:
            df = df.groupby(df.index).mean()

        print(f"Transcriptomics: Cleaned data shape: {df.shape}")
        return df
    except Exception as e:
        print(f"Error loading transcriptomics: {e}")
        return pd.DataFrame()

def load_epigenomics_features():
    path = FILE_MAP.get("epigenomics.soft")
    if not isinstance(path, str) or not (path.startswith("uploaded:") or os.path.exists(path)): return pd.DataFrame()
    try:
        # Epigenomics data often starts with '!' comments
        df = pd.read_csv(path, sep='\t', comment='!', engine='python', encoding='latin1')
        print(f"Epigenomics: Raw data shape: {df.shape}")
        id_col = 'Gene_ID'
        feature_col = 'Distance_to_TSS'

        if id_col not in df.columns or feature_col not in df.columns:
            print(f"Epigenomics: Missing critical columns ('{id_col}' or '{feature_col}').")
            return pd.DataFrame()

        df = df[[id_col, feature_col]].copy()
        # Clean Gene ID: remove 'GeneID:' prefix, strip whitespace, and ensure it's not empty
        df[id_col] = df[id_col].astype(str).str.replace(r'GeneID:', '', regex=True).str.strip()
        df = df.loc[df[id_col] != '']

        # CRITICAL FIX: Ensure the Gene IDs are numeric strings before setting index
        df['is_numeric'] = df[id_col].str.isnumeric()
        df = df[df['is_numeric']].drop(columns=['is_numeric'])
        df = df.set_index(id_col)

        df[feature_col] = pd.to_numeric(df[feature_col], errors='coerce')
        df.dropna(subset=[feature_col], inplace=True)

        # NEW FIX: Ensure unique index by aggregating by mean distance
        if not df.index.is_unique:
            df = df.groupby(df.index).mean()

        print(f"Epigenomics: Cleaned data shape: {df.shape}")
        return df
    except Exception as e:
        print(f"Error loading epigenomics: {e}")
        return pd.DataFrame()

def load_proteomics_features():
    path = FILE_MAP.get("proteomics.txt")
    if not isinstance(path, str) or not (path.startswith("uploaded:") or os.path.exists(path)): return pd.DataFrame()
    try:
        df = pd.read_csv(path, sep='\t', engine='python', encoding='latin1')
        print(f"Proteomics: Raw data shape: {df.shape}")
        uniprot_col = 'UniProt'
        npx_col = 'NPX'

        if uniprot_col not in df.columns or npx_col not in df.columns:
            print(f"Proteomics: Missing critical columns ('{uniprot_col}' or '{npx_col}').")
            return pd.DataFrame()

        # Group by UniProt ID and calculate the mean NPX value for each protein
        df[npx_col] = pd.to_numeric(df[npx_col], errors='coerce')
        df = df.groupby(uniprot_col)[npx_col].mean().to_frame() # Already ensures unique index

        df.index = df.index.astype(str).str.strip()
        df = df.loc[df.index != '']
        df.dropna(how='all', inplace=True)

        print(f"Proteomics: Cleaned data shape: {df.shape}")
        return df
    except Exception as e:
        print(f"Error loading proteomics: {e}")
        return pd.DataFrame()

# --- 3. OMICS FEATURE INTEGRATION ---
print("\n--- Generating Omics Features ---")
# Gene Features (Transcriptomics-gene, epigenomics-gene)
df_trans = load_transcriptomics_features()
df_epi = load_epigenomics_features()

gene_ids = set(df_trans.index) | set(df_epi.index)
df_gene_features = pd.DataFrame(index=pd.Index(list(gene_ids), name='Gene ID'))

if not df_trans.empty:
    # Ensure merge is safe even if there are different columns in the two omics sets
    df_gene_features = df_gene_features.merge(df_trans, left_index=True, right_index=True, how='left')
if not df_epi.empty:
    df_gene_features = df_gene_features.merge(df_epi, left_index=True, right_index=True, how='left')

df_gene_features.dropna(how='all', inplace=True)
gene_feature_cols = [col for col in df_gene_features.columns if df_gene_features[col].dtype in [np.float64, np.int64]]
df_gene_pca = process_features_with_pca(df_gene_features, gene_feature_cols)
print(f"Processed {df_gene_pca.shape[0]} gene nodes with {df_gene_pca.shape[1]} omics features.")

# Protein Features (proteomics-protein)
df_prot_features = load_proteomics_features()
protein_feature_cols = [col for col in df_prot_features.columns if df_prot_features[col].dtype in [np.float64, np.int64]]
df_prot_pca = process_features_with_pca(df_prot_features, protein_feature_cols)
print(f"Processed {df_prot_pca.shape[0]} protein nodes with {df_prot_pca.shape[1]} omics features.")

# Metabolite features loading skipped as requested
df_metabo_pca = pd.DataFrame()
print(f"Processed 0 metabolite nodes with 0 omics features (Skipped as requested).")

# --- 4. RE-MAP NODES AND LOAD EDGES ---
print("\n--- Re-Collecting All Node IDs (Including Omics IDs) ---")
# Update node sets with Omics IDs to ensure all are in the graph (although we will filter later)
all_nodes['gene'].update(df_gene_pca.index.tolist())
all_nodes['protein'].update(df_prot_pca.index.tolist())

# Re-create maps to include all new IDs
node_maps = {}
node_id_to_string = defaultdict(list)
for node_type, ids in all_nodes.items():
    clean_ids = {i for i in ids if i and i != 'nan'}
    categories = pd.Categorical(list(clean_ids)).categories
    node_maps[node_type] = pd.Series(np.arange(len(categories)), index=categories)
    node_id_to_string[node_type] = categories.tolist()
    if node_type != 'metabolite' or len(categories) > 0:
        print(f"Re-mapped {len(categories)} unique {node_type} nodes.")

# Initialize the heterogeneous graph object
data = HeteroData()

def load_and_encode_edges_generic(filename_key, type_src, type_dst, edge_rel, col_src, col_dst, sep):
    """Loads edges for a simple (src, dst) relationship and adds to HeteroData."""
    path = FILE_MAP.get(filename_key)
    is_uploaded = isinstance(path, str) and path.startswith("uploaded:")
    if not is_uploaded and (path is None or not os.path.exists(path)):
        # Do not print warning if the file was the one that previously failed column checks
        if "disease to gene-2 tsv" not in filename_key:
            print(f"Warning: File not found at {path}. Skipping edges for {filename_key}.")
        return
    try:
        # Use quoting=csv.QUOTE_NONE for TSV files
        quoting_mode = csv.QUOTE_NONE if sep == '\t' else csv.QUOTE_MINIMAL
        df = pd.read_csv(path, sep=sep, skiprows=1, header=None, comment='#', engine='python',
                         skipinitialspace=True, quoting=quoting_mode)

        # Robust check for column bounds
        max_col_index = max(col_src, col_dst)
        if max_col_index >= df.shape[1]:
            print(f"Error loading edges for {filename_key}: Columns {col_src} or {col_dst} out of bounds (DataFrame has {df.shape[1]} columns). Skipping.")
            return

        df[col_src] = df.iloc[:, col_src].astype(str).str.strip()
        df[col_dst] = df.iloc[:, col_dst].astype(str).str.strip()

        src_map = node_maps[type_src]
        dst_map = node_maps[type_dst]

        df = df[df[col_src].isin(src_map.index)]
        df = df[df[col_dst].isin(dst_map.index)]

        src_indices = src_map.loc[df[col_src]].values
        dst_indices = dst_map.loc[df[col_dst]].values

        edge_index = torch.tensor([src_indices, dst_indices], dtype=torch.long)

        # Only add to graph if there are edges to prevent PyG errors
        if edge_index.size(1) > 0:
            data[type_src, edge_rel, type_dst].edge_index = edge_index
            data[type_dst, f'rev_{edge_rel}', type_src].edge_index = edge_index.flip([0])
            print(f"Loaded {edge_index.size(1)} edges for ({type_src}, {edge_rel}, {type_dst}).")
        else:
            print(f"Skipped ({type_src}, {edge_rel}, {type_dst}): 0 edges found after mapping.")

    except Exception as e:
        print(f"Error loading edges for {filename_key}: {e}")

def load_and_encode_polypharmacy_se(filename_key, type_src, type_dst, edge_rel, col_d1, col_d2, col_se, sep):
    """Loads edges for polypharmacy (drug1, drug2, side_effect) -> drug-side_effect."""
    path = FILE_MAP.get(filename_key)
    is_uploaded = isinstance(path, str) and path.startswith("uploaded:")
    if not is_uploaded and (path is None or not os.path.exists(path)):
        return
    try:
        df = pd.read_csv(path, sep=sep, skiprows=1, header=None, comment='#', engine='python', skipinitialspace=True)
        max_col_index = max(col_d1, col_d2, col_se)
        if max_col_index >= df.shape[1]:
            print(f"Error loading polypharmacy SE edges for {filename_key}: Columns out of bounds (DataFrame has {df.shape[1]} columns). Skipping.")
            return

        df[col_d1] = df.iloc[:, col_d1].astype(str).str.strip()
        df[col_d2] = df.iloc[:, col_d2].astype(str).str.strip()
        df[col_se] = df.iloc[:, col_se].astype(str).str.strip()

        df_d1_se = df[[col_d1, col_se]].copy().rename(columns={col_d1: 'src_id', col_se: 'dst_id'})
        df_d2_se = df[[col_d2, col_se]].copy().rename(columns={col_d2: 'src_id', col_se: 'dst_id'})
        df_edges = pd.concat([df_d1_se, df_d2_se]).drop_duplicates()

        src_map = node_maps[type_src] # 'drug' map
        dst_map = node_maps[type_dst] # 'side_effect' map

        df_edges = df_edges[df_edges['src_id'].isin(src_map.index)]
        df_edges = df_edges[df_edges['dst_id'].isin(dst_map.index)]

        src_indices = src_map.loc[df_edges['src_id']].values
        dst_indices = dst_map.loc[df_edges['dst_id']].values

        edge_index = torch.tensor([src_indices, dst_indices], dtype=torch.long)

        if edge_index.size(1) > 0:
            data[type_src, edge_rel, type_dst].edge_index = edge_index
            data[type_dst, f'rev_{edge_rel}', type_src].edge_index = edge_index.flip([0])
            print(f"Loaded {edge_index.size(1)} edges for ({type_src}, {edge_rel}, {type_dst}) (Polypharmacy SE).")
        else:
            print(f"Skipped ({type_src}, {edge_rel}, {type_dst}) (Polypharmacy SE): 0 edges found after mapping.")

    except Exception as e:
        print(f"Error loading polypharmacy SE edges for {filename_key}: {e}")

def load_and_encode_polypharmacy_dd(filename_key, type_src, type_dst, edge_rel, col_d1, col_d2, col_se, sep):
    """Loads edges for polypharmacy (drug1, drug2, side_effect) -> drug-drug."""
    path = FILE_MAP.get(filename_key)
    is_uploaded = isinstance(path, str) and path.startswith("uploaded:")
    if not is_uploaded and (path is None or not os.path.exists(path)):
        return
    try:
        df = pd.read_csv(path, sep=sep, skiprows=1, header=None, comment='#', engine='python', skipinitialspace=True)
        max_col_index = max(col_d1, col_d2)
        if max_col_index >= df.shape[1]:
            print(f"Error loading polypharmacy DD edges for {filename_key}: Columns out of bounds (DataFrame has {df.shape[1]} columns). Skipping.")
            return

        df[col_d1] = df.iloc[:, col_d1].astype(str).str.strip()
        df[col_d2] = df.iloc[:, col_d2].astype(str).str.strip()
        df_edges = df[[col_d1, col_d2]].copy()
        df_edges.columns = ['src_id', 'dst_id']

        src_map = node_maps[type_src] # 'drug' map
        dst_map = node_maps[type_dst] # 'drug' map

        df_edges = df_edges[df_edges['src_id'].isin(src_map.index)]
        df_edges = df_edges[df_edges['dst_id'].isin(dst_map.index)]

        src_indices = src_map.loc[df_edges['src_id']].values
        dst_indices = dst_map.loc[df_edges['dst_id']].values

        edge_index = torch.tensor([src_indices, dst_indices], dtype=torch.long)

        if edge_index.size(1) > 0:
            data[type_src, edge_rel, type_dst].edge_index = edge_index
            data[type_dst, f'rev_{edge_rel}', type_src].edge_index = edge_index.flip([0])
            print(f"Loaded {edge_index.size(1)} edges for ({type_src}, {edge_rel}, {type_dst}) (Polypharmacy DD).")
        else:
             print(f"Skipped ({type_src}, {edge_rel}, {type_dst}) (Polypharmacy DD): 0 edges found after mapping.")

    except Exception as e:
        print(f"Error loading polypharmacy DD edges for {filename_key}: {e}")

# --- EXECUTE EDGE LOADING ---
print("\n--- Loading Network Edges (Using Final Maps) ---")
# Load standard and D-G_1, D-G_2 edges
for filename, type_src, type_dst, edge_rel, (col_src, col_dst, sep) in RELATION_DEFINITIONS:
    load_and_encode_edges_generic(filename, type_src, type_dst, edge_rel, col_src, col_dst, sep)

# Load Polypharmacy links
_, type_src, type_dst, edge_rel, (col_d1, col_d2, col_se, sep) = POLYPHARMACY_DRUG_SE
load_and_encode_polypharmacy_se(POLYPHARMACY_DRUG_SE[0], type_src, type_dst, edge_rel, col_d1, col_d2, col_se, sep)

_, type_src, type_dst, edge_rel, (col_d1, col_d2, col_se, sep) = POLYPHARMACY_DRUG_DRUG
load_and_encode_polypharmacy_dd(POLYPHARMACY_DRUG_DRUG[0], type_src, type_dst, edge_rel, col_d1, col_d2, col_se, sep)

# --- 5. FEATURE ASSIGNMENT ---
print("\n--- Assigning Omics and Learnable Features ---")
# Helper to assign features (either Omics PCA or Learnable Embeddings)
def assign_features(node_type, df_pca):
    num_nodes = len(node_maps[node_type])
    if num_nodes == 0:
        data[node_type].x = torch.empty((0, FEATURE_DIM)).to(device)
        return

    # Check if we have omics features and they match the feature dim
    if not df_pca.empty and df_pca.shape[1] == FEATURE_DIM and df_pca.shape[0] > 0:
        feature_matrix = np.zeros((num_nodes, FEATURE_DIM), dtype=np.float32)
        # Determine which Omics IDs also exist in the final graph map
        omics_ids = df_pca.index.intersection(node_maps[node_type].index)

        if not omics_ids.empty:
            # Get the index positions in the PyG graph for the matched Omics IDs
            indices = node_maps[node_type].loc[omics_ids].values
            # Get the features only for the matched IDs
            matched_features = df_pca.loc[omics_ids].values
            feature_matrix[indices] = matched_features

            # CRITICAL: If Omics features are present, they are FIXED features, not parameters.
            data[node_type].x = torch.tensor(feature_matrix, dtype=torch.float).to(device)
            print(f"Assigned {omics_ids.size} omics features (Fixed) to {node_type} ({df_pca.shape[1]} dims).")
            return

    # Fallback to learnable embedding
    # Initialize as parameter only if the node type actually exists
    if num_nodes > 0:
        data[node_type].x = torch.nn.Parameter(torch.Tensor(num_nodes, FEATURE_DIM))
        torch.nn.init.xavier_uniform_(data[node_type].x)
        data[node_type].x = data[node_type].x.to(device)
        print(f"Initialized {node_type} with {num_nodes} x {FEATURE_DIM} learnable features (Fallback).")
    else:
        # If no nodes exist for this type, still assign an empty tensor
        data[node_type].x = torch.empty((0, FEATURE_DIM)).to(device)
        print(f"Initialized {node_type} with 0 nodes.")

# Assign Features based on omics data presence (epigenomics-gene, Transcriptomics-gene, proteomics-protein)
assign_features('gene', df_gene_pca)
assign_features('protein', df_prot_pca)

# Assign learnable features for remaining node types (including empty 'metabolite')
for node_type in ['disease', 'drug', 'function', 'side_effect', 'metabolite']:
    assign_features(node_type, pd.DataFrame()) # Pass empty DF to force learnable fallback or empty tensor

# ----------------------------------------------------------------------
## MODEL ARCHITECTURE AND TRAINING SETUP
# ----------------------------------------------------------------------

class DeeperGraphSAGEEncoder(torch.nn.Module):
    """
    A deeper, memory-optimized GNN encoder using SAGEConv.
    Uses three layers for multi-hop neighborhood aggregation.
    """
    def __init__(self, metadata, hidden_channels, out_channels):
        super().__init__()

        # Helper to dynamically create conv layers for existing edge types
        def get_conv_dict(conv_layer, h_channels):
            conv_dict = {}
            for rel in metadata[1]:
                src, rel_type, dst = rel
                # Check if the edge type exists and has edges in the global 'data' object
                if (src, rel_type, dst) in data.edge_types and data[(src, rel_type, dst)].edge_index.numel() > 0:
                    # (-1, -1) for heterogeneous input features; SAGEConv uses mean aggregation by default
                    conv_dict[rel] = conv_layer((-1, -1), h_channels)
            return conv_dict

        # Three layers of SAGEConv for multi-hop neighborhood aggregation
        self.conv1 = HeteroConv(get_conv_dict(SAGEConv, hidden_channels), aggr='sum')
        self.conv2 = HeteroConv(get_conv_dict(SAGEConv, hidden_channels), aggr='sum')
        # Final layer outputs the desired dimension
        self.conv3 = HeteroConv(get_conv_dict(SAGEConv, out_channels), aggr='sum')

    def forward(self, x_dict, edge_index_dict):
        # Layer 1
        x_dict = self.conv1(x_dict, edge_index_dict)
        x_dict = {key: x.relu() for key, x in x_dict.items()}
        # Layer 2
        x_dict = self.conv2(x_dict, edge_index_dict)
        x_dict = {key: x.relu() for key, x in x_dict.items()}
        # Layer 3 (Output layer)
        x_dict = self.conv3(x_dict, edge_index_dict)
        return x_dict

class LinkPredictor(torch.nn.Module):
    """MLP classifier for link prediction using concatenated embeddings."""
    def __init__(self, in_channels):
        super().__init__()
        self.lin1 = torch.nn.Linear(2 * in_channels, in_channels)
        self.lin2 = torch.nn.Linear(in_channels, in_channels // 2)
        self.lin3 = torch.nn.Linear(in_channels // 2, 1)
        self.dropout = torch.nn.Dropout(0.3)
        self.bn = torch.nn.BatchNorm1d(in_channels)

    def forward(self, z_src, z_dst, edge_label_index):
        src_index, dst_index = edge_label_index
        # Concatenate source and destination embeddings
        z = torch.cat([z_src[src_index], z_dst[dst_index]], dim=-1)
        z = self.lin1(z)
        # BatchNorm expects (batch_size, num_features)
        self.bn.num_features = z.shape[1]
        z = self.bn(z)
        z = z.relu()
        z = self.dropout(z)
        z = self.lin2(z).relu()
        return self.lin3(z).squeeze()

# Target link for prediction: Disease-Gene Association 1
TARGET_EDGE = ('disease', 'associated_with_1', 'gene')
REV_TARGET_EDGE = ('gene', 'rev_associated_with_1', 'disease')

if TARGET_EDGE not in data.edge_types or data[TARGET_EDGE].edge_index.numel() == 0:
    print(f"\n--- Target Edge Missing ---")
    print(f"No valid target edges found for {TARGET_EDGE}. Aborting training.")
    # Exit cleanly if the target link for prediction is missing
    sys.exit()

# Split data for link prediction validation/testing
transform = RandomLinkSplit(
    num_val=0.1, num_test=0.1, neg_sampling_ratio=1.0,
    edge_types=[TARGET_EDGE], rev_edge_types=[REV_TARGET_EDGE]
)
train_data, val_data, test_data = transform(data)

# Move split data to device
train_data, val_data, test_data = train_data.to(device), val_data.to(device), test_data.to(device)

# Instantiate Model and Predictor
model = DeeperGraphSAGEEncoder(data.metadata(), hidden_channels=FEATURE_DIM, out_channels=FEATURE_DIM)
predictor = LinkPredictor(in_channels=FEATURE_DIM)
model, predictor = model.to(device), predictor.to(device)

# Collect all parameters that need optimization (includes model, predictor, and learnable node features)
learnable_params = (
    list(model.parameters()) +
    list(predictor.parameters()) +
    # Important: Collect only tensors that were defined as nn.Parameter (i.e., learnable embeddings)
    [param for x_dict in data.x_dict.values() for param in ([x_dict] if isinstance(x_dict, torch.nn.Parameter) else [])]
)

optimizer = torch.optim.Adam(learnable_params, lr=LEARNING_RATE)

# --- TRAINING, EVALUATION, AND PREDICTION FUNCTIONS ---

def train():
    model.train()
    predictor.train()
    optimizer.zero_grad()

    # 1. Forward pass through the encoder
    z_dict = model(train_data.x_dict, train_data.edge_index_dict)

    # 2. Get embeddings for the source and destination node types of the target edge
    z_src, z_dst = z_dict[TARGET_EDGE[0]], z_dict[TARGET_EDGE[2]]

    # 3. Prepare edge indices and labels for the predictor
    pos_edge_index = train_data[TARGET_EDGE].edge_label_index[:, train_data[TARGET_EDGE].edge_label == 1]
    neg_edge_index = train_data[TARGET_EDGE].edge_label_index[:, train_data[TARGET_EDGE].edge_label == 0]

    edge_label_index = torch.cat([pos_edge_index, neg_edge_index], dim=-1)
    edge_label = torch.cat([
        torch.ones(pos_edge_index.size(1)),
        torch.zeros(neg_edge_index.size(1))
    ]).to(device)

    # 4. Predict
    out = predictor(z_src, z_dst, edge_label_index)

    # 5. Compute loss
    loss = F.binary_cross_entropy_with_logits(out, edge_label.float())
    loss.backward()
    optimizer.step()
    return loss.item()

@torch.no_grad()
def evaluate(data_split):
    model.eval()
    predictor.eval()

    # Ensure all data is on the correct device before passing to model
    current_x_dict = {key: x.to(device) for key, x in data_split.x_dict.items()}
    current_edge_index_dict = {key: x.to(device) for key, x in data_split.edge_index_dict.items()}

    z_dict = model(current_x_dict, current_edge_index_dict)

    z_src, z_dst = z_dict[TARGET_EDGE[0]], z_dict[TARGET_EDGE[2]]

    edge_label_index = data_split[TARGET_EDGE].edge_label_index.to(device)
    edge_label = data_split[TARGET_EDGE].edge_label.to(device)

    out = predictor(z_src, z_dst, edge_label_index).view(-1)
    y_pred = torch.sigmoid(out).cpu()
    y_true = edge_label.float().cpu()

    if len(y_true.unique()) < 2:
        auc = 0.5
        ap = 0.0
    else:
        auc = roc_auc_score(y_true, y_pred)
        ap = average_precision_score(y_true, y_pred)
    return auc, ap



# --- MAIN EXECUTION ---
if __name__ == '__main__':
    print("\n--- Starting Model Training ---")
    best_val_auc = 0
    best_epoch = 0

    for epoch in range(1, NUM_EPOCHS + 1):
        loss = train()
        val_auc, val_ap = evaluate(val_data)
        test_auc, test_ap = evaluate(test_data)
        print(data.metadata())  # should now match your trained model


        if val_auc > best_val_auc:
            best_val_auc = val_auc
            best_epoch = epoch

            # Save the best model and predictor parameters
            torch.save(model.state_dict(), MODEL_ENCODER_PATH)
            torch.save(predictor.state_dict(), MODEL_PREDICTOR_PATH)

            # Save the learnable features (if any)
            for node_type in data.x_dict:
                if isinstance(data[node_type].x, torch.nn.Parameter):
                    torch.save(data[node_type].x.data.clone(), os.path.join(SAVE_DIR, f'best_learnable_feat_{node_type}.pt'))

            status = '(*)'
        else:
            status = ''

        print(f'Epoch: {epoch:03d}, Loss: {loss:.4f}, Val AUC: {val_auc:.4f}, Test AUC: {test_auc:.4f} {status}')
    print(f"\n--- Training Summary ---")
    print(f"Training finished. Best Validation AUC: {best_val_auc:.4f} achieved at Epoch {best_epoch}.")
    print(f"Model saved to {MODEL_ENCODER_PATH} and {MODEL_PREDICTOR_PATH}")
```
